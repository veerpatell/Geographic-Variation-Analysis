```python
import pandas as pd
```

```python
df = pd.read_csv("1.csv")
```

```python
services = [
    'IP_MDCR_STDZD_PYMT_PC',
    'IP_CVRD_STAYS_PER_1000_BENES',
    'OP_MDCR_STDZD_PYMT_PC',
    'OP_VISITS_PER_1000_BENES',
    'ASC_MDCR_STDZD_PYMT_PC',
    'ASC_EVENTS_PER_1000_BENES',
    'EM_MDCR_STDZD_PYMT_PC',
    'EM_EVNTS_PER_1000_BENES',
    'PRCDRS_MDCR_STDZD_PYMT_PC',
    'PRCDR_EVNTS_PER_1000_BENES',
    'TESTS_MDCR_STDZD_PYMT_PC',
    'TESTS_EVNTS_PER_1000_BENES',
    'IMGNG_MDCR_STDZD_PYMT_PC',
    'IMGNG_EVNTS_PER_1000_BENES'
]
```

```python
# Calculate year-over-year percentage changes for each service
for service in services:
    df[f'Percentage Change {service}'] = ((df[service] - df[service].shift(1)) / df[service].shift(1)) * 100
```
